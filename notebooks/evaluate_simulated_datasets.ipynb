```python
import pandas as pd
from sklearn.metrics import roc_auc_score
```

```python
df = pd.read_csv('../data/eas/eas_dataset_fingerprint_xtb_simulated_whole.csv')
df
```

```python
source_df = df[df['simulation_idx'] == 0]

for idx in [1, 2, 3, 4]:
    sim_df = df[df['simulation_idx'] == idx]
    
    preds, targets = [], []
    for reaction_idx in sim_df['reaction_idx']:
        pred = sim_df[sim_df['reaction_idx'] == reaction_idx]['label'].values[0]
        target = source_df[source_df['reaction_idx'] == reaction_idx]['label'].values[0]
        preds.append(pred)
        targets.append(target)
    
    print(len(sim_df))
    print(roc_auc_score(targets, preds))
```

```python
import pandas as pd
from sklearn.metrics import roc_auc_score
df = pd.read_csv('../data/da/global_simulated_DA_literature/dataset.csv')
df
```

```python
# compute scores
targets, preds = [], []
right = 0
for idx in df['reaction_idx'].unique():
    selection = df[df['reaction_idx'] == idx]
    target = selection[selection['simulation_idx'] == 0]['label']
    pred = selection[selection['simulation_idx'] == 1]['label']

    if len(pred) > 0 and len(target) > 0 and 1 in pred.values:

        if (target.values == pred.values).all():
            right += 1

        for val in target:
            targets.append(val)
        for val in pred:
            preds.append(val)

print(roc_auc_score(targets, preds) * 100, 'AUROC')
print(right/ len(df['reaction_idx'].unique()) * 100, "%", "accuracy")
```
